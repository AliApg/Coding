# some information about DataFrames

```python
# seattle_weather stores information about the weather in Seattle, 
# and austin_weather stores information about the weather in Austin. 
# Each of the data frames has a MONTHS column that stores the three-letter name of the months. 
# Each also has a column named "MLY-PRCP-NORMAL" that stores the average rainfall in each month during a ten-year period.
```

```python
# Create a Figure and an Axes with plt.subplots by ggplot style sheets.
plt.style.use('ggplot')
fig, ax = plt.subplots()

# Plot MLY-PRCP-NORMAL from seattle_weather against the MONTH so that:
# 1-Its color be blue.
# 2-The marker points be a circle.
# 3-The linestyle be continuous dashes.

ax.plot(sea["M"], sea["MLY"], color='b', marker='o', linestyle='--', label='seattle')

# Plot MLY-PRCP-NORMAL from austin_weather against MONTH so that:
# 1-Its color be red.
# 2-The marker points be a triangle_down.
# 3-The linestyle be continuous dashe-dot.

ax.plot(aus["M"], aus["MLY"], color='r', marker='v', linestyle='-.', label='austin')

# Customize the x-axis label: "Time (months)".
ax.set_xlabel("Time (months)")

# Customize the y-axis label: "Precipitation (inches)".
ax.set_ylabel("Precipitation (inches)")

# Add the title: "Weather patterns in Austin and Seattle".
ax.set_title("Weather patterns in Austin and Seattle")

# Add the legend without any frame with 2 column in the upper center of the figure
ax.legend(frameon=False, loc='upper center', ncol=2)

# Call the show function
plt.show()
```

# Small multiples with shared y axis

```python
# Create a figure and an array of axes: 
# 1 rows, 2 column with shared y axis. 
# the figure size should be 30 cm in length and 10 cm in width.
#Plot Seattle's "MLY-PRCP-NORMAL" in a solid blue line in the top Axes.
#Add Seattle's "MLY-PRCP-25PCTL" and "MLY-PRCP-75PCTL" in dashed blue lines to the top Axes.

#Plot Austin's "MLY-PRCP-NORMAL" in a solid red line in the top Axes and the "MLY-PRCP-25PCTL" and "MLY-PRCP-75PCTL" in dashed red lines.
```

```python
def cm_to_inch(value):
    return value/2.54
```

```python
plt.style.use('default')
fig, ax = plt.subplots(1, 2, sharey=True, figsize=(cm_to_inch(30), cm_to_inch(10)))


# The first chart from left:
# Plot Seattle's "MLY-PRCP-NORMAL"(precipitation) against MONTH in a solid blue line
# Add Seattle's "MLY-PRCP-25PCTL" and "MLY-PRCP-75PCTL" in dashed blue lines

ax[0].plot(sea["M"], sea["MLY"], color = 'b')
ax[0].plot(sea["M"], sea["MLY25"], color = 'b', linestyle = '--')
ax[0].plot(sea["M"], sea["MLY75"], color = 'b', linestyle = '--')

# The second chart from left:
# Plot austin's "MLY-PRCP-NORMAL"(precipitation) against MONTH in a solid blue line
# Add austin's "MLY-PRCP-25PCTL" and "MLY-PRCP-75PCTL" in dashed blue lines.

ax[1].plot(aus['M'], aus['MLY'], color = 'r')
ax[1].plot(aus['M'], aus['MLY25'], color = 'r', linestyle = '--')
ax[1].plot(aus['M'], aus['MLY75'], color = 'r', linestyle = '--')

# Customize the y-axis label for each chart: "Precipitation (inches)".
ax[0].set_ylabel("Precipitation (inches)")
ax[1].set_ylabel("Precipitation (inches)")

# Add the title for each chart: "Weather patterns in Austin and Seattle".
ax[0].set_title("Weather patterns in Austin and Seattle")
ax[1].set_title("Weather patterns in Austin and Seattle")

plt.show()
```

```python
# *Import climate_change CSV file as a DataFrame: climate
```

```python
climate = pd.read_csv('climate_change.csv', parse_dates=["date"], index_col="date")
print(climate)
```

```python
# 1-Use plt.subplots to create a Figure and Axes objects called fig and ax, respectively.
# 2-Plot the carbon dioxide variable in blue using the Axes plot method.
# 3-Use the Axes twinx method to create a twin Axes that shares the x-axis.
# 4-Plot the relative temperature variable in the twin Axes using its plot method.
```

```python
# 1-Initalize a Figure and Axes
fig, ax = plt.subplots()

# 2-Plot the CO2 variable in blue
ax.plot(climate.index, climate["co2"], color="blue")

# Customize the y-axis label: "CO2 (ppm)".
ax.set_ylabel("CO2 (ppm)")
ax.tick_params("y", colors="blue")

# 3-Create a twin Axes that shares the x-axis
ax2 = ax.twinx()

# 4-Plot the relative temperature in red
ax2.plot(climate.index, climate["relative_temp"], color="red")

# Customize the x-axis label: "Time (months)".
ax.set_xlabel("Time")
ax2.tick_params("y", colors="red")


# Customize the y-axis label: "Precipitation (inches)".
ax2.set_ylabel("Reletive temperature (Celsius)")

plt.show()
```

```python
# Import Olympic medals CSV file as a DataFrame: medals
```

```python
medals = pd.read_csv('medals_by_country_2016.csv', index_col=0)
print(medals)
```

```python
# Call the ax.bar method to plot the "Gold" column as a function of the country.
# Use the ax.set_xticklabels to set the x-axis tick labels to be the country names.
# In the call to ax.set_xticklabels rotate the x-axis tick labels by 90 degrees by using the rotation key-word argument.
# Set the y-axis label to "Number of medals".
```

```python
fig, ax = plt.subplots()

# Plot a bar-chart of gold medals as a function of country
ax.bar(medals.index, medals["Gold"])

# Set the x-axis tick labels to the country names
ax.set_xticklabels(medals.index, rotation=90)

# Add bars for "Gold" with the label "Gold"
ax.bar(medals.index, medals["Gold"], label="Gold")

# Stack bars for "Silver" on top with label "Silver"
ax.bar(medals.index, medals["Silver"], bottom= medals["Gold"], label="Silver")

# Stack bars for "Bronze" on top of that with label "Bronze"
ax.bar(medals.index, medals["Bronze"], bottom=medals["Gold"]+medals["Silver"], label = "Bronze")

# Display the legend
ax.legend()

# Set the y-axis label
ax.set_ylabel("Number of medals")

plt.show()
```

# Plot a pie chart.

```python
# Call the ax.pie method to plot the the persent of "Gold", "Silver", and "Bronze" medals in the US.
# Draw a shadow beneath the pie.
# Calculate the percent of each wedge and report.
# Provide the labels for each wedge: labels
# Specify 0.1 radius with which to offset "Gold" wedge.
# Specify thhe angle of 90 by which the start of the pie is rotated, counterclockwise from the x-axis.
```

## By default the plotting of the first wedge starts from the x-axis and move counterclockwise:

```python
medals = pd.read_csv('medals_by_country_2016.csv', index_col=0)
print(medals)
```

```python
# 1-proportion in 1_D array.
sizes=medals.iloc[[0],[0,1,2]]
proportion=sizes.to_numpy().squeeze()

# 2-labels.
labels = list(medals.columns)

# 3-offsetting a slice with "explode".
explode = (0.1, 0, 0) 

# 4-Plot a pie-chart of US medals in 2016.

fig, ax = plt.subplots()

ax.pie(proportion, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True,  startangle=90)

# 5-Equal aspect ratio ensures that pie is drawn as a circle.
ax.axis('equal')

# 6-Customize the title label: "US Olympic Medals in 2016.".
ax.set_title("US Olympic Medals in 2016.")

# 7-Display the legend
ax.legend(bbox_to_anchor=(.5, .5), loc=3)

plt.show()
```

```python
gdp = pd.read_csv('gdp.csv')
print(gdp)
```

```python
# Create a scatter plot: 
# 1 rows, 5 column with shared y axis. 
# the figure size should be 40 cm in length and 10 cm in width.
# set the x and y axes in logarithmic scale.
# Plot Seattle's "gdpPercap" over "pop" in each continent.
# Use the "year" of each observation as input to color each point according to its date.
# Use the "lifeExp" of each observation as input to size of each point according to its date.
```

```python
def cm_to_inch(value):
    return value/2.54
```

```python
gdp["country"]=gdp["country"].astype('category')
```

```python
fig, ax = plt.subplots(1,5,sharey=True, figsize=(cm_to_inch(40), cm_to_inch(10)))

# Asia
ax[0].scatter(gdp.groupby("continent").get_group("Asia")["pop"], gdp.groupby('continent').get_group('Asia')["gdpPercap"],
             c=gdp.groupby("continent").get_group("Asia")["year"],
             s=gdp.groupby("continent").get_group("Asia")["lifeExp"])

ax[0].set_xscale('log')
ax[0].set_yscale('log')
ax[0].set_title('Asia')
ax[0].set_xlabel("Population")
ax[0].set_ylabel("gdpPercap")

# Africa
ax[1].scatter(gdp.groupby("continent").get_group("Africa")["pop"], gdp.groupby('continent').get_group('Africa')["gdpPercap"],
             c=gdp.groupby("continent").get_group("Africa")["year"],
             s=gdp.groupby("continent").get_group("Africa")["lifeExp"])

ax[1].set_xscale('log')
ax[1].set_yscale('log')
ax[1].set_title('Africa')
ax[1].set_xlabel("Population")

# Americas
ax[2].scatter(gdp.groupby("continent").get_group("Americas")["pop"], gdp.groupby('continent').get_group('Americas')["gdpPercap"],
             c=gdp.groupby("continent").get_group("Americas")["year"],
             s=gdp.groupby("continent").get_group("Americas")["lifeExp"])

ax[2].set_xscale('log')
ax[2].set_yscale('log')
ax[2].set_title('Americas')
ax[2].set_xlabel("Population")

# Europe
ax[3].scatter(gdp.groupby("continent").get_group("Europe")["pop"], gdp.groupby('continent').get_group('Europe')["gdpPercap"],
             c=gdp.groupby("continent").get_group("Europe")["year"],
             s=gdp.groupby("continent").get_group("Europe")["lifeExp"])

ax[3].set_xscale('log')
ax[3].set_yscale('log')
ax[3].set_title('Europe')
ax[3].set_xlabel("Population")

# Oceania
ax[4].scatter(gdp.groupby("continent").get_group("Oceania")["pop"], gdp.groupby('continent').get_group('Oceania')["gdpPercap"],
             c=gdp.groupby("continent").get_group("Oceania")["year"],
             s=gdp.groupby("continent").get_group("Oceania")["lifeExp"])

ax[4].set_xscale('log')
ax[4].set_yscale('log')
ax[4].set_title('Oceania')
ax[4].set_xlabel("Population")

plt.show()
```
